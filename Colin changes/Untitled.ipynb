```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("commodity_trade_statistics_data.csv", low_memory=False)
```

```python
data
```

```python
data.groupby('country_or_area')


```

```python
def getData(country, commodity, title):
    all_data = data[(data['commodity']==commodity) & (data['year']>=1995) & (data['country_or_area']==country) & (data['flow']=='Import')]
    year_and_quantity_data = {'Year': all_data['year'], 'Quantity': all_data['quantity']}
    df = pd.DataFrame(year_and_quantity_data, columns= ['Year', 'Quantity'])
    df.reset_index(drop=True, inplace=True)
    x_axis = df['Year']
    y_axis = df['Quantity']
    plt.plot(x_axis, y_axis, label=country)
    plt.title(title)
    plt.xlabel("Year")
    plt.ylabel("Quantity x 100,000,000")
```

```python
getData("USA", 'Sugar confectionery not chewing gum, no cocoa content', "Confectionary Imports With No Chocolate")
getData("China", 'Sugar confectionery not chewing gum, no cocoa content', "Confectionary Imports With No Chocolate")
getData("Russian Federation", 'Sugar confectionery not chewing gum, no cocoa content', "Confectionary Imports With No Chocolate")
getData("Germany", 'Sugar confectionery not chewing gum, no cocoa content', "Confectionary Imports With No Chocolate")
getData("Canada", 'Sugar confectionery not chewing gum, no cocoa content', "Confectionary Imports With No Chocolate")
plt.legend(loc="upper right")
```

```python
getData("USA", 'Chocolate and other food preps containing cocoa > 2 k', "Chocolate Foods Imports")
getData("China", 'Chocolate and other food preps containing cocoa > 2 k', "Chocolate Foods Imports")
getData("Russian Federation", 'Chocolate and other food preps containing cocoa > 2 k', "Chocolate Foods Imports")
getData("Germany", 'Chocolate and other food preps containing cocoa > 2 k', "Chocolate Foods Imports")
getData("Canada", 'Chocolate and other food preps containing cocoa > 2 k', "Chocolate Foods Imports")
plt.legend(loc="upper right")
```
